```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("titanic data.csv")
```

```python

print(df.isnull().sum())
```

```python
# Fill missing Age with median
df['Age'].fillna(df['Age'].median(), inplace=True)

# Fill missing Fare with median
df['Fare'].fillna(df['Fare'].median(), inplace=True)

# Drop 'Cabin' due to too many missing values
df.drop(columns=['Cabin'], inplace=True)

# Confirm cleaning
print(df.isnull().sum())
```

** Univariate Analysis**

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Age'], bins=30, kde=True, color='skyblue')
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

```python
sns.countplot(x='Sex', data=df, palette='Set2')
plt.title("Gender Distribution")
plt.show()
```

```python
sns.countplot(x='Pclass', data=df, palette='Set3')
plt.title("Passenger Class Distribution")
plt.show()
```

**Bivariate Analysis**

```python
sns.countplot(x='Survived', hue='Sex', data=df, palette='Set1')
plt.title("Survival by Gender")
plt.legend(title="Sex", loc="upper right")
plt.show()
```

```python
sns.countplot(x='Survived', hue='Pclass', data=df, palette='Set2')
plt.title("Survival by Passenger Class")
plt.legend(title="Class", loc="upper right")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Survived', y='Age', data=df, palette='Pastel1')
plt.title("Age vs Survival")
plt.show()
```

**Correlation Heatmap**

```python
plt.figure(figsize=(10,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Between Numerical Variables')
plt.show()
```
